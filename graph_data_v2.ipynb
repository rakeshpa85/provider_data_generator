```python
tin_per_state_cnt = 1
group_tin_seed = 123456789
prov_cnt_rand_min = 1
prov_cnt_rand_max = 100
provider_npi_seed = 100000000
addr_cnt_rand_min = 1
addr_cnt_rand_max = 10

prov_mpin_seed = 1000
fake = Faker()

random.seed(101)
prov_detail_list = []
avbl_st_list=['AL','AK','AZ','AR','CA','CO','CT','DC','FL','GA','KY','MD','MA','OK','TN','VT']
spec_code_len = len(list(spec_df['SUBSPECIALTY_CODE']))


for i in tqdm (range (len(avbl_st_list)), desc="Loading…", ascii=False, ncols=75):
    for tin_cnt in range(1,(tin_per_state_cnt+1)):
#         print(i, group_tin_seed)
        group_tin_seed += 1

        prov_cnt_rand = random.randint(prov_cnt_rand_min, prov_cnt_rand_max)
        for prov_cnt in range(1,(prov_cnt_rand+1)):
#             print(i, group_tin_seed,provider_npi_seed)
            provider_npi_seed += 1
            prov_name = fake.name()

            addr_cnt_rand = random.randint(addr_cnt_rand_min, addr_cnt_rand_max)
            for addr_cnt in range(1,(addr_cnt_rand+1)):
#                 print(i, group_tin_seed,provider_npi_seed)
                addr_dict = random_address.real_random_address_by_state(avbl_st_list[i])
                if len(addr_dict) == 6:
#                     print(addr_dict)
                    prov_addr = addr_dict['address1'] + " " + addr_dict['address2'] + " " + addr_dict['city'] + " " + addr_dict['state'] + " " + addr_dict['postalCode']
                    prov_cord = (addr_dict['coordinates']['lat'], addr_dict['coordinates']['lng'])
                    rand_subspec_code = random.randint(1, spec_code_len)
                    spec_det_list = list(spec_df[spec_df['SUBSPECIALTY_CODE'] == rand_subspec_code].iloc[0])
                    prov_detail_list.append([group_tin_seed,provider_npi_seed,prov_name,prov_mpin_seed,prov_addr,prov_cord]+spec_det_list)
                    prov_mpin_seed += 1
                    
prov_detail_df = pd.DataFrame(prov_detail_list,columns=['group_tin','prov_npi','prov_name','prov_mpin','prov_addr','prov_cord','spec_code','subspec_code','spec','subspec'])
prov_detail_df.shape
```
